```python
import pandas as pd
```

```python
df_test = pd.read_csv('https://raw.githubusercontent.com/odubno/Bike-Sharing-Demand/master/test.csv')
df_train = pd.read_csv('https://raw.githubusercontent.com/odubno/Bike-Sharing-Demand/master/train.csv')
```

```python
from datetime import datetime
import numpy as np

np.set_printoptions(threshold=np.nan)

def transform(df):
    i = 0
    for timestamp in df['datetime']:
        i += 1
        date_object = datetime.strptime(timestamp.split()[0], '%Y-%m-%d')
        date = datetime.date(date_object).weekday()
        time = timestamp.split()[1][:2]
        df.loc[i-1, 'date'] = date
        df.loc[i-1, 'time'] = time
    return df
```

```python
train, test = transform(df_train), transform(df_train)
```

```python
import pylab as pl
import numpy as np
from sklearn.ensemble import RandomForestRegressor

X = ['date','time','season','holiday','workingday','weather','temp','atemp','humidity','windspeed']

clf = RandomForestRegressor(n_estimators = 200)

model_casual = clf.fit(train[X], train.casual)

print model_casual.feature_importances_
predict_casual = clf.predict(test[X])

model_registered = clf.fit(train[X], train.registered)

print model_registered.feature_importances_
predict_registered = clf.predict(test[X])
```
